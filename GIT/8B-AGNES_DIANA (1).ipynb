```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df = df.apply(encoder.fit_transform)
df
```

```python
x = df.drop(['Class'],axis =1)
y = df['Class']

sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title("AGNES")
```

```python
cluster = AgglomerativeClustering(n_clusters= 3,affinity= 'euclidean',linkage='ward')
cluster.fit(x)
labels = cluster.labels_
labels
```

```python
x = np.array(x)
plt.scatter(x[labels == 0,0],x[labels == 0,1],s=100,color = 'red')
plt.scatter(x[labels == 1,0],x[labels == 1,1],s=100,color = 'green')
plt.scatter(x[labels == 2,0],x[labels == 2,1],s=100,color = 'blue')
plt.title("Agnes")
plt.show()
```

```python
x = df.drop(['Class'],axis =1)
y = df['Class']

sch.dendrogram(sch.linkage(x,method = 'complete'))
plt.title("DIANA")
```

```python
cluster = AgglomerativeClustering(n_clusters= 3,affinity= 'euclidean',linkage='complete')
cluster.fit(x)
labels = cluster.labels_
labels
```

```python
x = np.array(x)
plt.scatter(x[labels == 0,0],x[labels == 0,1],s=100,color = 'red')
plt.scatter(x[labels == 1,0],x[labels == 1,1],s=100,color = 'green')
plt.scatter(x[labels == 2,0],x[labels == 2,1],s=100,color = 'blue')
plt.title("DIANA")
plt.show()
```
